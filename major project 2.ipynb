```python
from nltk.corpus import wordnet as wn
from nltk.stem.wordnet import WordNetLemmatizer
from collections import defaultdict

#This function is used to Find the root form of the worst
def lemma(tokens):
	tag_map = defaultdict(lambda : wn.NOUN)
	tag_map['J'] = wn.ADJ
	tag_map['V'] = wn.VERB
	tag_map['R'] = wn.ADV

	not_list = ['JJR', 'JJS', 'RBR', 'RBS']

	lemma_function = WordNetLemmatizer()
	list = []
	
	for token, tag in tokens:
		if tag in not_list: #We dont wanna lemmatize comparitive and superlative terms 
			list.append(tuple([token,tag])) #Such terms are kept without any alteration
		else:
			list.append(tuple([( lemma_function.lemmatize(token, tag_map[tag[0]]) ),tag]))
	
	return (list)
```

```python
import os
import timeit 
from symspellpy.symspellpy import SymSpell, Verbosity  # import the module

#This module loads the dictionary in reference to which the spellings are corrected
#This is called only once at the initiation of the process and not repetedly 
class checker:
    def __init__(self):
        max_edit_distance_dictionary = 2
        prefix_length = 7
        self.sym_spell = SymSpell(max_edit_distance_dictionary, prefix_length)
   
        dictionary_path = os.path.join(os.path.dirname(__file__),
                                   "frequency_dictionary_en_82_765.txt")
        term_index = 0  
        count_index = 1  
        if not self.sym_spell.load_dictionary(dictionary_path, term_index, count_index):
            print("Dictionary file not found")

#This module Checks for the best possible spelling correction substitute if present
    def spellcorrect(self , s):
        input_term = (s)
        max_edit_distance_lookup = 2 #so that not to go vague in correcting the terms
        suggestions = self.sym_spell.lookup_compound(input_term,
                                                max_edit_distance_lookup)
        
        for suggestion in suggestions:
            return(suggestion.term)
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
  
def stp_wrd(word_tokens):
    stop_words = set(stopwords.words('english')) #Loading the list of stop words we created

    filtered_sentence = [] 
    
    #To check if our word is present in stop words list
    for token,tag in word_tokens: 
        if token not in stop_words: 
            filtered_sentence.append(tuple([token, tag])) 

    return filtered_sentence
```

```python
#This program calls all the functions 
from nltk.corpus import stopwords
from Symspell import checker
import Lemmatization as lm
import Abbreviation_Expansion as ab
import Stop_wrd as sw
import POS_Tagging as pt
import Score as sc
import timeit
import nltk
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize

#To load spelling dictionary before hand and common for all commments, which other wise will take time
t1 = timeit.default_timer()
ob1 = checker()
t2 = timeit.default_timer()
print ('time to load dictionary ',t2-t1)

def sentiment_score(s, n): #here n=1 for sentiment analysis and n=0 for pre-processing only
    ##this is not score it

    if n == 1:
        s = s.replace( '\n' , '.')

    #This is to call function which expands all abbreviations and slangs and texting terms
    #t1 = timeit.default_timer()
        abb_s = ab.abbrev(s) #return type string


    #Calling the module to lower case the entire string for standardization
    if n ==1:
        abb_s = abb_s.lower()

    #To remove any common greeting terms from the text. 
    if n == 1:
        abb_s.replace('good day', '')  
        abb_s.replace('good night', '')
        abb_s.replace('good morning', '')
        abb_s.replace('nice day', '')

    #To tokenize the text in sentences and send them for any speeling correction
    list = sent_tokenize(abb_s)

    #Calling the module to correct if any spelling mistake
    if n == 1:
        correct_s = ''
        for item in list:
                correct_s = correct_s + ob1.spellcorrect(item) + '. ' #return type string
   
   #This step is to attain the original Part of speech (POS) tag of words in the sentence
   #This is necessary before any deletion in the data is made
    tokens = word_tokenize(correct_s)
    tagged = nltk.pos_tag(tokens)

    #This step in to remove all the unwanted POS-tags
    filtered_s = pt.tagIT(tagged) #return type list with tupples

    #Calling the module to Lemmatize the text and bring it in any root form if posiible
    if n == 1:
        lemma_s = lm.lemma( filtered_s ) #return type list with tupples

    #Calling the module to remove stop words
    stp_s = sw.stp_wrd(lemma_s)  #return type is list with tuples
    #print (stp_s)

    if n == 0:
        accept = ['NN', 'NNS', 'NNP', 'NNPS', 'VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ' ]
        string = ''
        for token, tag in stp_s:
                if tag in accept:
                        string += token + ' '
        return (string)


    #Calling the module to find the final score 
    cmpo, dict = sc.scoreIT(stp_s)
    #t2 = timeit.default_timer()
    #time = t2-t1
    return (cmpo, dict)
```
